```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

n = 500
values = [0, 1, 2, 3, 4]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.1]

# Ensure the probabilities sum to 1
probabilities = np.array(probabilities) / sum(probabilities)

children_counts = np.random.choice(values, size=n, p=probabilities)
unique, counts = np.unique(children_counts, return_counts=True)
pmf = counts / n
cumulative_probabilities = np.cumsum(pmf)

# Plot PMF in a separate graph
plt.figure(figsize=(8, 4))
sns.barplot(x=unique, y=pmf, palette="viridis")
plt.xlabel("Number of Children Per Family")
plt.ylabel("Probability Mass Function (PMF)")
plt.title("Probability Mass Function (PMF) of Number of Children")
plt.ylim(0, max(pmf) * 1.1)
plt.show()

# Plot CDF in a separate graph
plt.figure(figsize=(8, 4))
sns.lineplot(x=unique, y=cumulative_probabilities, marker="o", color="blue")
plt.xlabel("Number of Children Per Family")
plt.ylabel("Cumulative Probability (CDF)")
plt.title("Cumulative Distribution Function (CDF) of Number of Children")
plt.ylim(0, 1.1)
plt.show()
```
